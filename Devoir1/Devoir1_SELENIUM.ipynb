```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time


# Path to chromedriver.exe on Windows
chromedriver_path = r'D:\\chromedriver\\chromedriver.exe'  # <-- update this path if needed

# Setup Chrome options
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Optional: run without opening a browser window
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Setup Service with the path to chromedriver
service = Service(executable_path=chromedriver_path)

# Initialize driver
driver = webdriver.Chrome(service=service, options=options)
```

```python
# Use the driver
driver.get('https://fstt.ac.ma/Portail2023/category/articles/actualites/')
```

```python
titles = []
links = []
dates = []

while True :
    time.sleep(1)
    articles = driver.find_elements(By.TAG_NAME,"article")
    for article in articles :
        titles.append(article.find_element(By.TAG_NAME,'h3').text)
        links.append(article.find_element(By.TAG_NAME,'h3').find_element(By.TAG_NAME,'a').get_attribute("href"))
        dates.append(article.find_element(By.TAG_NAME,'span').text)

    try:
        next_button = driver.find_element(By.CLASS_NAME, 'next')  # or use a better selector
        if 'disabled' in next_button.get_attribute('class'):
            break
        next_button.click()
    except:
        break
```

```python
print(titles)
print(links)
print(dates)
```

```python
# Create folder to save images
import os
import requests

os.makedirs('downloaded_images', exist_ok=True)
driver.get('https://fstt.ac.ma/Portail2023/')

# Get all <img> elements
images = driver.find_elements(By.TAG_NAME, 'img')

for idx, img in enumerate(images):
    src = img.get_attribute('src') or img.get_attribute('data-src')
    if not src:
        continue

    try:
        img_data = requests.get(src).content
        # Save image with unique name
        with open(f'downloaded_images_devoir1/image_{idx}.jpg', 'wb') as f:
            f.write(img_data)
        print(f"Downloaded: image_{idx}.jpg")
    except Exception as e:
        print(f"Failed to download {src}: {e}")
```

```python
# Close driver
driver.quit()
```

```python
import pandas as pd

df = pd.DataFrame({
    'Title': titles,
    'Link': links,
    'Date': dates
})

df.to_csv("Devoir1_SELENIUM.csv",index=False)
df
```
